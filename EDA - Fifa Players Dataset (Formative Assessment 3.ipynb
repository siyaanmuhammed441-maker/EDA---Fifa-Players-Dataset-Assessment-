```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("fifa_data.csv")

top_country = df['Nationality'].value_counts().idxmax()
print("Country with most players:", top_country)

top_5_countries = df['Nationality'].value_counts().head(5)

plt.figure()
top_5_countries.plot(kind='bar')
plt.title("Top 5 Countries with Most FIFA Players")
plt.xlabel("Country")
plt.ylabel("Number of Players")
plt.show()

df['Wage'] = df['Wage'].replace('[K]', '', regex=True).astype(float)

highest_paid_player = df.loc[df['Wage'].idxmax(), 'Name']
print("Highest paid player:", highest_paid_player)

plt.figure()
plt.hist(df['Wage'], bins=30)
plt.title("Salary Distribution of FIFA Players")
plt.xlabel("Wage (in K)")
plt.ylabel("Number of Players")
plt.show()

def height_to_cm(height):
    feet, inches = height.split("'")
    return int(feet) * 30.48 + int(inches.replace('"', '')) * 2.54

df['Height_cm'] = df['Height'].apply(height_to_cm)

tallest_player = df.loc[df['Height_cm'].idxmax(), 'Name']
print("Tallest player:", tallest_player)

top_club = df['Club'].value_counts().idxmax()
print("Club with most players:", top_club)

preferred_foot = df['Preferred Foot'].value_counts().idxmax()
print("Most preferred foot:", preferred_foot)

plt.figure()
df['Preferred Foot'].value_counts().plot(kind='bar')
plt.title("Preferred Foot of FIFA Players")
plt.xlabel("Foot")
plt.ylabel("Number of Players")
plt.show()
```
